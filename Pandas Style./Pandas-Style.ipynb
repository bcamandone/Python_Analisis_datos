```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("u.data.txt",sep="\t",header=None)
```

```python
data1.columns = ["user_id","item_id","rating","timestamp"]
```

```python
data1 = data1.drop("timestamp", axis=1)
```

```python
data1.head()
```

```python
data2 = data1.iloc[:20]
```

```python
data2.shape
```

```python
headers = {
    'selector': 'th.col_heading',
    'props': 'background-color: #5E17EB; color: white;'
}

index_style = {
    'selector': 'th.index_name',
    'props': 'background-color: #5E17EB; color: white;'
}

data2_style = (
    data2
        .style
            .set_table_styles([headers,index_style])
            .set_properties(**{'background-color': '#FFCFC9','color': 'black'})
)

data2_style
```

```python

columns = data2.columns[data2.columns.str.startswith('rating')]

max_value = data2[columns].max().max()
min_value = data2[columns].min().min()
# Establecer el estilo para la celda con el valor m√°ximo
max_style =  f'background-color:#3BE8B0; '    
# Establecer el estilo para la celda con el valor m√≠nimo
min_style = f'background-color: #FF66C4; '
(
 data2
 .style
 .set_table_styles([headers])
 .applymap(lambda x: max_style if x == max_value else '')
 .applymap(lambda x: min_style if x == min_value else '', subset=columns)
)
```

```python
(
    data2
        .style
            .set_table_styles([headers, index_style])
            .background_gradient(cmap='Reds',subset=columns)
)
```
