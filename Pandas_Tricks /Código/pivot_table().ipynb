## Función pivot_table()

```python
import pandas as pd
import numpy as np
fortune = pd.read_csv("Fortune.csv")
fortune.head()
```

```python
mean_Profits = fortune["revenue"].mean() # calculo el promedio 
fortune["Revenue_segmentation"] = np.where(fortune["revenue"] > mean_Profits, "High", "Low")
```

```python
fortune["Revenue_segmentation"].value_counts()
```

```python
pd.pivot_table(data = fortune, 
               index = ['sector'],
               columns = ["Revenue_segmentation"],
               aggfunc = 'size',
              fill_value = 0) # 'size' indica que se está contando el número de ocurrencias (tamaño) 
```

```python
# Crear la tabla pivotada
pivot_table = pd.pivot_table(data = fortune,
                             values = 'revenue', 
                             index = ['sector','city'],
                             columns = ["Revenue_segmentation"],
                             aggfunc = {"revenue": ["sum", 'size']}, 
                             fill_value = 0)


pivot_table.columns = ['_'.join(col).strip() for col in pivot_table.columns]


pivot_table
```
