## Visualizar datos faltantes con missingno

missingno es una libreria de Python excelente y f치cil de usar, que proporciona una serie de visualizaciones para comprender la presencia y distribuci칩n de datos faltantes. Esto puede ser en forma de gr치fico de barras, gr치fico de matriz, mapa de calor o dendrograma.

```python
import pandas as pd
import  missingno as msno
```

```python
data = pd.read_csv("titanic_train.csv") 
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
msno.bar(data)
```

En el lado izquierdo de la gr치fica, la escala del eje Y var칤a de 0,0 a 1,0, donde 1,0 representa el 100 % de la integridad de los datos. Si la barra es menor que esto, indica que nos faltan valores dentro de esa columna.

```python
msno.matrix(data)
```

Como se ve en el gr치fico resultante, las columnas Age,Cabin, y Embarked muestran datos faltantes. Esto se identific칩 en el diagrama de barras, pero el beneficio adicional es que podemos ver c칩mo se distribuyen los datos que faltan en el dataframe.

```python
msno.heatmap(data)
```

El mapa de calor se utiliza para identificar correlaciones de nulidad entre cada una de las diferentes columnas. 
Los valores cercanos a 1 positivo indican que la presencia de valores nulos en una columna est치 correlacionada con la presencia de valores nulos en otra columna.
Los valores cercanos a 0 indican que hay poca o ninguna relaci칩n entre la presencia de valores nulos en una columna en comparaci칩n con otra.

```python
msno.dendrogram(data)
```
