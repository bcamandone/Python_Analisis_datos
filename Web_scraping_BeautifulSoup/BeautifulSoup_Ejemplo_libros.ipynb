```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://books.toscrape.com/"

response = requests.get(url) 
if response.status_code == 200: # Compruebar si la solicitud se realizó correctamente
    print("Solicitud exitosa")
else:
    print("Solicitud fallida")
```

Analizar el contenido HTML usando BeautifulSoup, buscar los datos que queremos extraer de la página.

```python
soup = BeautifulSoup(response.text, "html.parser") 
print(soup)
```

```python

books = soup.find_all("article", class_="product_pod")

# Iterar a través de los libros y extraer de cada libro, su titulo, precio, rating y stock
for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").get_text().strip()
    rating = book.p["class"][1]
    stock = book.find("p", class_="instock availability").get_text().strip()
    
    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Rating: {rating}")
    print(f"Available: {stock}")
```
